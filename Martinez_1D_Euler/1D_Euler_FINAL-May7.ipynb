## Principles of the Shock Tube Problem

Initially, the tube is divided (by a diaphragm) into a high-pressure section (driver section) and a low pressure section (the driven section). There is no flow velocity anywhere inside the tube initially. 

As, the diaphragm is removed, the initial pressure discontinuity will propagate to the right as an unsteady normal shock wave traveling at velocity W. At the same time, an isentropic expansion wave will propagate to the left.

We can see that shortly after the diaphragm is removed, there are 4 main regions in the tube.

Region 1 is the undisturbed portion of the driven section (so we would expect initial conditions to still be true here)

Region 2 :  This section has been ‘processed’ by the shock wave moving through it, and the pressure here is equal to the pressure behind the shock.

Region 3: Has been ‘processed’ by the expansion wave moving through the region. The pressure here is equal to P3. Regions 2 and 3 are said to be at the same pressure since the gas in these regions cannot support any pressure discontinuities

Region 4: The undisturbed portion of the driver section (we should expect initial conditions).

It is interesting that Regions 2 and 3 are at the same velocity and pressure. But there are differences. Since region 2 has been processed by a shock wave and region 3 by an expansion wave, the entropy, temperature, and density are different. This is why we see this contact surface to differentiate the regions.

Both Pressure and velocity change discontinuously across the shock wave. When we look at these properties across the expansion wave, we see that the variations of pressure and velocity are finite and continuous.

This information was taken from Anderson(1995).

## Governing Equations

We look at one-dimensional conservation of mass, momentum, and energy:

$$\frac{\partial \rho}{\partial t} + \frac{\partial \rho \vec{v_x}}{\partial x} = 0$$

$$\frac{\partial \rho \vec{v_x}}{\partial t} + \frac{\partial(\rho \vec{v_x}^2 +P)}{\partial x} = 0$$

$$\frac{\partial E}{\partial t} +\ \frac{\partial (E+P) \vec{v_x}}{\partial x} = 0$$

We can also relate an equation of state:

$$ P = (\gamma-1)(E - \frac{\rho v^2}{2})$$

## System of Equations

From the conservation Equations we can develop a system of equations:

$$\frac{\partial U}{\partial t} + \frac{\partial f(U)}{\partial x} = 0$$

where U and f(U) are:

$$ U = \left(
\begin{array}{c}
\rho\\
\rho \vec{v_x}\\
E
\end{array}
\right)
\,\
f(U) = \left(
\begin{array}{c}
\rho \vec{v_x}\\
\rho \vec{v_x}^2 +\ P\\
(E+P)\vec{v_x}
\end{array}
\right)$$ 

## Discretization of System of Equations: The Discontinuous Galerkin Method

The discontinuous Galerkin method represents unknowns like the finite element method by piecewise polynomial functions, but unlike finite element methods, the polynomials are discontinuous at cell interfaces. At these interfaces, a flux is defined in the same way as for finite volume methods [Sonnendrucker, 2013. Numerical Methods for Hyperbolic Systems]

On each cell, the unknown U is represented by a linear combination of basis functions of the space of polynomials of degree 'np'. The chosen basis function for this solver is a Legendre polynomial, and the interpolation points are Gauss-Lobatto points.

$$\int_{I_i} \frac{\partial U(x,t)}{\partial t} W(x)\ dx\ + \int_{I_i} \frac{\partial f(U(x,t))}{\partial x} W(x) dx = 0$$

$$\int_{I_i} \frac{\partial U(x,t)}{\partial t} \phi_l(x)\ dx\ + \int_{I_i} \frac{\partial f(U(x,t))}{\partial x} \phi_l(x) dx = 0$$

$$\int_{I_i} \frac{\partial U(x,t)}{\partial t} \phi_l(x)\ dx\ + f(U)\phi_l(x)\ |_{I_i}\ -\ \int_{I_i} f(U)\frac{\partial \phi_l(x)}{\partial x} dx = 0$$


Note that the Integral I goes from [-0.5] t0 [0.5]. From the Galerkin Method we can estimate U(x,t):

$$U_{i}(x,t) = \sum_{0}^{np} U_{l}^{i}(t)\phi_l(x)$$

The objective here is to obtain an expression for U(t). Using orthogonality, we can define the spectral coefficients, which in turn allow us to define U(t):

$$\alpha_l = \int_{I_i} \phi_l(x)\ \phi_l(x) dx\\$$
$$U_{l}^{i}(t) = \frac{1}{\alpha_l}\ \int_{I_i} U_{i}(x,t)\ \phi_l(x) dx\\$$

Then, we can express the system of equations in a form that we can approximate:

$$\frac{\partial U_{l}^{i}(t)}{\partial t} = \frac{1}{\alpha_l}  
\int_{I_i} f(U)\frac{\partial \phi_l(x)}{\partial x} dx 
- f(U_{i+\frac{1}{2}})\phi_{i+\frac{1}{2}}^l 
+ f(U_{i-\frac{1}{2}})\phi_{i-\frac{1}{2}}^l $$

The integral on the right-hand side can be approximated by Gauss-Lobatto quadratures and the fluxes are solved using the Rusanov flux. To march in time, a third-order runge-kutta method is used.

The solver is tested by applying it to the classic Sod's shock tube problem. It is executed using a third order polynomial approximation, with a domain of 10 cells.

### Some Program Parameters

```python
np = 3 # order of scheme
ncell = 10 #number of cells
nface = ncell+1 #cell boundaries
rk3 = 3 #This is because we are solving 3 equations(mass, momentum, energy)
dt = 0.01

# for flux function:
normal = 1.0
```

## Third Order Runge Kutta Time Scheme

##### Stages of RK Scheme:



$$U^1 = U^n\ +\ \Delta R(U^n) \\$$
$$U^2 = \frac{3}{4}U^n\ +\ \frac{1}{4}\ [U^1\ +\ \Delta R(U^1)] \\$$
$$U^3 = \frac{1}{3}U^n\ +\ \frac{2}{3}\ [U^2\ +\ \Delta R(U^2)] \\$$


##### Initialize Runge-Kutta Scheme:

```python
Q_rk = U_initial
```

```python
Qf = get_Qf(ncell, rk3, np, fle_gauss, Q_rk)
```

### Boundary Conditions

```python
Qbleft, Qbright = get_Qbright_Qbleft(rk3,np, ncell, Qf)
```

##### First Cell Left Edge

```python
Qfl, Qfr = get_Qfl_Qfr_left(rk3, Qf, np, Qbleft)
```

```python
Fl = get_rusanov_flux(Qfl, Qfr, normal, rk3)
```

```python
bflux = get_bflux_left(Fl)
```

##### Last Cell Right Edge

```python
Qfl = get_Qfl_right(Qf,np, ncell)
```

```python
Fl = get_rusanov_flux(Qfl, Qbright, normal, rk3)
```

```python
bflux = get_bflux_right(rk3, ncell, Fl)
```

##### Interior Edge Flux

```python
bflux = get_bflux_interior(nface, f2c, Qf, np, get_rusanov_flux)
```

### All Volume Flux

```python
volflux = get_volflux(rk3, np, ncell, Qf)
```

### Initialize h matrix

```python
h = get_initial_hmatrix(rk3, np, ncell, bflux, fle_gauss)
```

### Volume Inegral by Gauss Lobatto

```python
h = get_GL_hmatrix(rk3, np, ncell, h, volflux,volcoef, fle_grad_gauss)
```

### Compute The Residual Function

```python
residual = get_residual_matrix(rk3, np, ncell, aic, h, dx)
```

## Runge-Kutta Stages 1 through 3 for Initial Time

```python
Qrk1 = get_Qrk_stage1(Q_rk, ncell, rk3, np, dt, residual)

# Pre-Stage 2 RK, follow the steps as above to obtain new residual 
```

```python
Qf = get_Qf(ncell, rk3, np, fle_gauss, Qrk1)

#BC

Qbleft, Qbright = get_Qbright_Qbleft(rk3,np, ncell, Qf)

#1st cell left edge

Qfl, Qfr = get_Qfl_Qfr_left(rk3, Qf, np, Qbleft)
Fl = get_rusanov_flux(Qfl, Qfr, normal, rk3)
bflux = get_bflux_left(Fl)

#last cell right edge

Qfl = get_Qfl_right(Qf,np, ncell)
Fl = get_rusanov_flux(Qfl, Qbright, normal, rk3)
bflux = get_bflux_right(rk3, ncell, Fl)

#Interior Edge Flux

bflux = get_bflux_interior(nface, f2c, Qf, np, get_rusanov_flux)
volflux = get_volflux(rk3, np, ncell, Qf)

#Initialize h matrix

h = get_initial_hmatrix(rk3, np, ncell, bflux, fle_gauss)

# Volume Flux GL

h = get_GL_hmatrix(rk3, np, ncell, h, volflux,volcoef, fle_grad_gauss)

#Compute residual (mass matrix)

residual = get_residual_matrix(rk3, np, ncell, aic, h, dx)

#Stage 2 Rk--------------------------------------------------------------------

Qrk2 = get_Qrk_stage2(Q_rk, Qrk1, ncell, rk3, np, dt, residual)

#Pre Stage 3 Rk-----------------------------------------------------------------

Qf = get_Qf(ncell, rk3, np, fle_gauss, Qrk2)

#BC

Qbleft, Qbright = get_Qbright_Qbleft(rk3,np, ncell, Qf)

#1st cell left edge

Qfl, Qfr = get_Qfl_Qfr_left(rk3, Qf, np, Qbleft)
Fl = get_rusanov_flux(Qfl, Qfr, normal, rk3)
bflux = get_bflux_left(Fl)

#last cell right edge

Qfl = get_Qfl_right(Qf,np, ncell)
Fl = get_rusanov_flux(Qfl, Qbright, normal, rk3)
bflux = get_bflux_right(rk3, ncell, Fl)

#Interior Edge Flux

bflux = get_bflux_interior(nface, f2c, Qf, np, get_rusanov_flux)
volflux = get_volflux(rk3, np, ncell, Qf)

#Initialize h matrix

h = get_initial_hmatrix(rk3, np, ncell, bflux, fle_gauss)

# Volume Flux GL

h = get_GL_hmatrix(rk3, np, ncell, h, volflux,volcoef, fle_grad_gauss)

#Compute residual (mass matrix)

residual = get_residual_matrix(rk3, np, ncell, aic, h, dx)

#Stage 3 Rk-------------------------------------------------------------

Qrk3= get_Qrk_stage3(Q_rk, Qrk2, ncell, rk3, np, dt, residual)
```
